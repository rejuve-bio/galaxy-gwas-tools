```python
import pandas as pd

# Read the gzipped file
df = pd.read_csv(
    "/mnt/hdd_1/ofgeha/galaxytool/Data/21001_raw.gwas.imputed_v3.both_sexes.tsv.bgz",
    sep="\t",
    compression="gzip"
)

# Split variant column into CHR, POS, REF, ALT
df[['CHR', 'POS', 'REF', 'ALT']] = df['variant'].str.split(":", expand=True)

# Convert POS to integer
df['POS'] = df['POS'].astype(int)

# Now pass the DataFrame directly to GWASLab
my_sumstats = gl.Sumstats(
    df,
    snpid="variant",
    chrom="CHR",
    pos="POS",
    ea="minor_allele",
    beta="beta",
    se="se",
    p="pval",
    n="n_complete_samples",
    maf="minor_AF",
    verbose=True
)

# Now liftover should work
my_sumstats.liftover(from_build="19", to_build="38")
```
