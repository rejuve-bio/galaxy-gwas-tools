Import Libraries and Define Paths

```python
# Import necessary Python libraries
import os
import pandas as pd

# ----------------- Configuration Paths -----------------
# Base directory of the GWAS harmonizer project
BASE_DIR = "/mnt/hdd_1/ofgeha/gwas-sumstats-harmoniser"

# Input GWAS summary statistics file
SUMSTATS = "/mnt/hdd_1/ofgeha/gwas-sumstats-harmoniser/Data/21001_raw.gwas.imputed_v3.both_sexes.tsv.bgz"

# Reference genome build
BUILD = "GRCh37"

# Directory containing reference GWAS data for harmonization
REF_DIR = "/mnt/hdd_1/ofgeha/gwas-sumstats-harmoniser/Data/gwas-ref"

# Path to the code repository (used for scripts and workflows)
CODE_REPO = BASE_DIR

# Allele frequency matching threshold for harmonization
THRESHOLD = "0.99"

# ----------------- Export as Environment Variables -----------------
# These variables will be accessible by Bash scripts and Nextflow
os.environ["SUMSTATS"] = SUMSTATS
os.environ["BUILD"] = BUILD
os.environ["REF_DIR"] = REF_DIR
os.environ["CODE_REPO"] = CODE_REPO
os.environ["THRESHOLD"] = THRESHOLD

# Print confirmation
print("Parameters exported:")
for key in ["SUMSTATS", "BUILD", "REF_DIR", "CODE_REPO", "THRESHOLD"]:
    print(f"{key} = {os.environ[key]}")
```
