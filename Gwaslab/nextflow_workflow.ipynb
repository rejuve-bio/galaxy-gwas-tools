Import Libraries and Define Paths

```python
import os
import pandas as pd
```

```python


# ---- paths based on your ls output ----
BASE_DIR = "/mnt/hdd_1/ofgeha/gwas-sumstats-harmoniser"

SUMSTATS = "/mnt/hdd_1/ofgeha/gwas-sumstats-harmoniser/Data/21001_raw.gwas.imputed_v3.both_sexes.tsv.bgz"
BUILD = "GRCh37"
REF_DIR = "/mnt/hdd_1/ofgeha/gwas-sumstats-harmoniser/Data/gwas-ref"
CODE_REPO = BASE_DIR
THRESHOLD = "0.99"

# ---- export for bash / nextflow ----
os.environ["SUMSTATS"] = SUMSTATS
os.environ["BUILD"] = BUILD
os.environ["REF_DIR"] = REF_DIR
os.environ["CODE_REPO"] = CODE_REPO
os.environ["THRESHOLD"] = THRESHOLD

print("Parameters exported:")
for k in ["SUMSTATS", "BUILD", "REF_DIR", "CODE_REPO", "THRESHOLD"]:
    print(f"{k}={os.environ[k]}")
```

Run Harmonization Workflow

```python
%%bash
set -euo pipefail

echo "Running harmonizer with:"
echo "  SUMSTATS   = $SUMSTATS"
echo "  BUILD      = $BUILD"
echo "  REF_DIR    = $REF_DIR"
echo "  CODE_REPO  = $CODE_REPO"
echo "  THRESHOLD  = $THRESHOLD"
echo

cd "$CODE_REPO"

./harmonizer.sh \
  --input "$SUMSTATS" \
  --build "$BUILD" \
  --ref "$REF_DIR" \
  --threshold "$THRESHOLD" \
  --code-repo "$CODE_REPO"
```
