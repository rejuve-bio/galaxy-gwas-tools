```python
import pandas as pd
import gwaslab as gl

# Load gzipped GWAS summary stats
df = pd.read_csv(
    "/mnt/hdd_1/ofgeha/galaxytool/Data/21001_raw.gwas.imputed_v3.both_sexes.tsv.bgz",
    sep="\t",
    compression="gzip"
)

# Split variant into CHR, POS, REF, ALT
df[['CHR', 'POS', 'REF', 'ALT']] = df['variant'].str.split(":", expand=True)
df['POS'] = df['POS'].astype(int)
```

```python
mysumstats = gl.Sumstats(df,fmt="gwaslab",other=["NOTE"])
```

```python
mysumstats.data
```

```python
gl.check_available_ref()
```

```python
gl.download_ref("1kg_dbsnp151_hg19_auto")
```

```python
# Rename minor_AF to EAF
mysumstats.data.rename(columns={"minor_AF": "EAF"}, inplace=True)

# Now harmonize
mysumstats.harmonize(basic_check=True,  # Perform basic check
                     threads=3,         # Number of threads to use
                     ref_seq="/mnt/hdd_1/ofgeha/galaxytool/Data/ref/hg19.fa",  # Reference genome
                     ref_infer="/mnt/hdd_1/ofgeha/galaxytool/Data/ref/EAS.ALL.split_norm_af.1kgp3v5.hg19.vcf.gz",  # Reference VCF file
                     ref_alt_freq="AF")  # Field for alternative allele frequency in the VCF
```
